```python
import pandas as pd
import numpy as np
import re
```

```python
file = './data/BMC-ContainmentZones.js'
```


**Regex to parse similar pattern to following code snnipet**
~~~js
var marker_9795626cfd584471ab4406d756a00baf = L.marker(
    [19.041691972000024, 72.85052482000003],
    {}
).addTo(feature_group_ad623471194f451d9f1cf7fc718747c5);
~~~

it will be able to get the value for marker id, long, lat data 



```python
data =None
data  = open(file,'r').read()
""" read data """
with open(file,'r') as f:
    data = f.read()
f.close()

""" parser """

pattern = """var[\t ]*marker_[0-9a-zA-Z]+[\n\t ]*=[\n\t ]*L.marker[\n\t ]*\([\n\t ]*\[[\n\t ]*[0-9]*[.][0-9]+"""
pattern+="""[\n\t ]*[,][\n\t ]*[0-9]*[.][0-9]+[\n\t ]*\]"""

match = re.findall(pattern =pattern,string = data)
print('Total Match', len(match))

""" randomly check 3 pattern match"""
n = np.random.randint(1,len(match))
print('\nCheck random sample\n',match[n:n+3])


""" Store All marker, latitue logitude"""
marker_list = []
latitude_longitude_pair = []

for string in match:
    marker_list.append(re.findall(pattern='_[0-9a-z]+',string=string)[0][1:])
    latitude_longitude_pair.append(re.findall(pattern='[0-9]*[.][0-9]+',string = string))



print(f'\n Total marker :: {len(marker_list)} \n Check random  :: \
      {marker_list[np.random.randint(1,len(marker_list))]}')


print(f'\n Total latitude_longitude_pair :: {len(latitude_longitude_pair)} \n Check random  :: \
      {latitude_longitude_pair[np.random.randint(1,len(latitude_longitude_pair))]}')


latitude, longitude = [i[0] for i in latitude_longitude_pair],[i[1] for i in latitude_longitude_pair]

""" Create DataFrame"""
df =pd.DataFrame(data ={'marker_id':marker_list,
                    'latitude':latitude,
                    'longitude':longitude
                   })
```
